```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X,y=make_blobs(n_samples=150,n_features=2,centers=2,cluster_std=1.05,random_state=42)

def scatter(X,y):
    plt.scatter(X[:,0][y==0],X[:,1][y==0],color="r")
    plt.scatter(X[:,0][y==1],X[:,1][y==1],color="b")
    plt.title("Scatter plot")
```

```python
scatter(X,y)
```

```python
class perceptron:
    def __init__(self):
        self.weights = np.zeros(2)
        self.bias=1
        self.learning_rate = 0.01
        
    def predict(self, inputs):
        predicted_value = np.dot(self.weights, inputs) + self.bias
        return 1 if predicted_value > 0 else 0
    
    def fit(self, training_inputs, labels, epochs = 100):
        for _ in range(epochs):
            for inputs, label in zip(training_inputs, labels):
                predicted = self.predict(inputs)
                error = label - predicted
                self.weights += self.learning_rate*error*np.array(inputs)
                self.bias += self.learning_rate*error
        
```

```python
model=perceptron()
model.fit(X,y,epochs=20)
```

```python
w=model.weights
b=model.bias

x1=[min(X[:,0]),max(X[:,0])]
x2=-(b+np.array(w[0])*x1)/np.array(w[1])
```

```python
scatter(X,y)
plt.plot(x1,x2)
```
